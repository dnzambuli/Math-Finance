```python
def euler_method(f, t0, y0, h, num_steps):
    """
    Euler method to solve the ODE y' = f(t, y), y(t0) = y0.
    
    Parameters:
    f : function
        The function f(t, y) representing the right-hand side of the ODE.
    t0 : float
        Initial time.
    y0 : float
        Initial value of y at t0.
    h : float
        Step size.
    num_steps : int
        Number of steps to compute.
        
    Returns:
    t_values : list
        List of time values.
    y_values : list
        List of corresponding y values.
    """
    t_values = [t0]
    y_values = [y0]
    
    t = t0
    y = y0
    
    for _ in range(num_steps):
        y += h * f(t, y)
        t += h
        
        t_values.append(t)
        y_values.append(y)
    
    return t_values, y_values

# Define the function f(t, y) = -2ty
def f(t, y):
    return -2 * t * y

# Initial conditions
t0 = 0
y0 = 1
h = 0.2
num_steps = int(1 / h)  # Number of steps to cover interval [0, 1]

# Apply Euler method
t_values, y_values = euler_method(f, t0, y0, h, num_steps)

# Print the results
for t, y in zip(t_values, y_values):
    print(f"t = {t:.2f}, y = {y:.6f}")
```

```python
def heun_method(f, t0, y0, h, num_steps):
    """
    Heun's method (Improved Euler method) to solve the ODE y' = f(t, y), y(t0) = y0.
    
    Parameters:
    f : function
        The function f(t, y) representing the right-hand side of the ODE.
    t0 : float
        Initial time.
    y0 : float
        Initial value of y at t0.
    h : float
        Step size.
    num_steps : int
        Number of steps to compute.
        
    Returns:
    t_values : list
        List of time values.
    y_values : list
        List of corresponding y values.
    """
    t_values = [t0]
    y_values = [y0]
    
    t = t0
    y = y0
    
    for _ in range(num_steps):
        # Predictor step (Euler's method)
        y_pred = y + h * f(t, y)
        
        # Corrector step
        y = y + 0.5 * h * (f(t, y) + f(t + h, y_pred))
        
        t += h
        
        t_values.append(t)
        y_values.append(y)
    
    return t_values, y_values

# Define the function f(t, y) = -2ty
def f(t, y):
    return -2 * t * y

# Initial conditions
t0 = 0
y0 = 1
h = 0.2
num_steps = int(1 / h)  # Number of steps to cover interval [0, 1]

# Apply Heun's method
t_values, y_values = heun_method(f, t0, y0, h, num_steps)

# Print the results
for t, y in zip(t_values, y_values):
    print(f"t = {t:.2f}, y = {y:.6f}")
```

```python
def rk4_method(f, t0, y0, h, num_steps):
    """
    Fourth-order Runge-Kutta method to solve the ODE y' = f(t, y), y(t0) = y0.
    
    Parameters:
    f : function
        The function f(t, y) representing the right-hand side of the ODE.
    t0 : float
        Initial time.
    y0 : float
        Initial value of y at t0.
    h : float
        Step size.
    num_steps : int
        Number of steps to compute.
        
    Returns:
    t_values : list
        List of time values.
    y_values : list
        List of corresponding y values.
    """
    t_values = [t0]
    y_values = [y0]
    
    t = t0
    y = y0
    
    for _ in range(num_steps):
        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)
        k3 = h * f(t + h/2, y + k2/2)
        k4 = h * f(t + h, y + k3)
        
        y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        t += h
        
        t_values.append(t)
        y_values.append(y)
    
    return t_values, y_values

# Define the function f(t, y) = -2ty
def f(t, y):
    return -2 * t * y

# Initial conditions
t0 = 0
y0 = 1
h = 0.2
num_steps = int(1 / h)  # Number of steps to cover interval [0, 1]

# Apply RK4 method
t_values, y_values = rk4_method(f, t0, y0, h, num_steps)

# Print the results
for t, y in zip(t_values, y_values):
    print(f"t = {t:.2f}, y = {y:.6f}")
```

```python
#finite difference method for heat equation 
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1.0  # Length of the rod
T = 0.5  # Total simulation time
alpha = 0.01  # Thermal diffusivity
N = 100  # Number of spatial points
M = 2000  # Number of time steps

# Spatial and temporal step sizes
dx = L / N
dt = T / M

# Initialize temperature array
u = np.zeros((N+1, M+1))

# Initial condition: u(x, 0) = sin(pi*x)
x = np.linspace(0, L, N+1)
u[:, 0] = np.sin(np.pi * x)

# Boundary conditions: u(0, t) = 0, u(L, t) = 0
u[0, :] = 0  # u(0, t) = 0
u[-1, :] = 0  # u(L, t) = 0

# Finite difference method
for n in range(M):
    for i in range(1, N):  # Exclude boundaries
        u[i, n+1] = u[i, n] + alpha * dt / (dx**2) * (u[i+1, n] - 2*u[i, n] + u[i-1, n])

# Plotting
plt.figure(figsize=(10, 6))
for n in range(0, M+1, int(M/10)):  # Plot every 10th timestep
    plt.plot(x, u[:, n], label=f't = {n * dt:.2f}')
plt.title('Temperature Distribution over Time')
plt.xlabel('Position x')
plt.ylabel('Temperature u(x, t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Monte Carlo estimation for pi
import random
import math

def monte_carlo_pi_estimation(num_samples):
    inside_circle = 0

    for _ in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)

        if math.sqrt(x**2 + y**2) <= 1:
            inside_circle += 1

    pi_estimate = (inside_circle / num_samples) * 4
    return pi_estimate

# Example usage:
num_samples = 1000000
pi_estimate = monte_carlo_pi_estimation(num_samples)
print(f"Estimated value of Pi with {num_samples} samples: {pi_estimate.numerical_approx(100)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_heat_equation(nx, ny, nt, alpha, dt, initial_temp, boundary_temp):
    # Initialize temperature grid
    T = np.zeros((nx, ny))
    T[:, :] = initial_temp
    T[0, :] = T[-1, :] = boundary_temp  # Top and bottom boundary conditions
    T[:, 0] = T[:, -1] = boundary_temp  # Left and right boundary conditions

    # Time integration
    for _ in range(nt):
        T_new = T.copy()
        for i in range(1, nx-1):
            for j in range(1, ny-1):
                T_new[i, j] = T[i, j] + alpha * dt * (
                    (T[i+1, j] - 2*T[i, j] + T[i-1, j]) / (dx**2) +
                    (T[i, j+1] - 2*T[i, j] + T[i, j-1]) / (dy**2)
                )
        T = T_new.copy()

    return T

# Parameters
nx, ny = 50, 50        # Grid size
nt = 500               # Number of time steps
alpha = 0.01           # Thermal diffusivity
dt = 0.1               # Time step size
dx, dy = 1.0, 1.0      # Grid spacing
initial_temp = 20.0    # Initial temperature
boundary_temp = 100.0  # Boundary temperature

# Solve heat equation
temperature_distribution = monte_carlo_heat_equation(nx, ny, nt, alpha, dt, initial_temp, boundary_temp)

# Plot the result
plt.imshow(temperature_distribution, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Temperature Distribution')
plt.show()
```
