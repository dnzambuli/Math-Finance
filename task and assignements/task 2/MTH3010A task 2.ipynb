## 4. Write a python code to check your results and plot all the numerical solutions (the 4. three) along with the exact solution in one graph.

````python
def euler_method(f, y0, t0, t_end, h):
    N = int((t_end - t0) / h)
    t = t0
    y = y0
    
    t_values = [t]
    y_values = [y]
    
    for _ in range(N):
        y += h * f(t, y)  
        t += h
        t_values.append(t)
        y_values.append(y)
    
    return t_values, y_values
````

````python
def f(t, y):
    return t - y
````

````python
# set up 
y0 = 1
t0 = 0
t_end = 2 # time end
h = 0.1
````

````python
euler_t, euler_y = euler_method(f, y0, t0, t_end, h)
````

## Euler output

````python
for t, y in zip(euler_t, euler_y):
    print(f"t = {t:.1f}, y = {y:.4f}")
````

````python
def heun_method(f, y0, t0, t_end, h):
    N = int((t_end - t0) / h)
    t = t0
    y = y0

    t_values = [t]
    y_values = [y]

    for _ in range(N):
        y1 = y + h * f(t, y) # yn + hf(tn, yn)
    
        y = y + (h / 2) * (f(t, y) + f(t + h, y1))
        t += h
        
        t_values.append(t)
        y_values.append(y)
    
    return t_values, y_values
````

````python
heun_t, heun_y = heun_method(f, y0, t0, t_end, h)
````

````python
for t, y in zip(heun_t, heun_y):
    print(f"t = {t:.1f}, y = {y:.5f}")
````

````python
def Runge_method(f , y0, t0, t_end, h):
    N = int((t_end - t0) / h)
    t = t0
    y = y0

    t_values = [t]
    y_values = [y]
    k1_val = []
    k2_val = []
    k3_val = []
    k4_val = []
    for _ in range(N):
        k1 = f(t, y)
        k2 = f(t + 0.5*h, y + 0.5 * h * k1)
        k3 = f(t + 0.5*h, y + 0.5 * h * k2)
        k4 = f(t + h, y + h * k3)

        y = y + h/6 * (k1 + 2*k2 + 2*k3 + k4)
        t += h

        t_values.append(t)
        y_values.append(y)
        k1_val.append(k1)
        k2_val.append(k2)
        k3_val.append(k3)
        k4_val.append(k4)

    return t_values, y_values, k1_val, k2_val, k3_val, k4_val
        
````

````python
Runge_t, Runge_y, Runge_k1, Runge_k2, Runge_k3, Runge_k4 = Runge_method(f, y0, t0, t_end, h)
````

````python
for t, y in zip(Runge_t, Runge_y):
    print(f"t = {t:.1f}, y = {y:.5f}")
````
